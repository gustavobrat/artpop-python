## ARTPOP Activation: Sessão 2

Contexto:

A *Haus of Gaga* está medindo a energia criativa de diferentes projetos. Cada tipo de projeto gera uma quantidade específica de **G.U.Y. (Girl Under You) points**:

* Projeto de Escultura (baseado em Jeff Koons): 10 pontos G.U.Y.
* Projeto de Performance (baseado em Marina Abramović): 8 pontos G.U.Y.
* Projeto de Moda (baseado em Versace): 6 pontos G.U.Y.

**Problema:**

Gaga supervisionou um projeto de cada categoria. Agora ela quer saber quantos G.U.Y. points acumulou com sua arte.

Nosso objetivo é criar um programa em Python que:

* Armazene a pontuação de cada categoria de projeto em uma variável.
* Calcule o total de pontos G.U.Y. acumulados por Gaga.
* Mostre na tela o total de pontos G.U.Y. acumulados.

```python
ponto_escultura = 10
ponto_performance = 8
ponto_moda = 6

pontos_gaga = ponto_escultura + ponto_performance + ponto_moda

print(f"Os pontos G.U.Y. totais de Gaga são: {pontos_gaga}")
```

## ARTPOP Activation: Sessão 3


**Divisão de Fichas de Poker (Poker Face, mas agora é Jewels n' Drugs)**

A equipe da *Haus of Gaga* encontrou um saco de fichas de poker (ou seriam jóias e drogas?). Para garantir uma divisão justa (segundo as regras do álbum), todos concordaram com a sugestão de Gaga, a líder da equipe:

Cada membro da equipe, exceto Gaga, deve receber exatamente o mesmo número de fichas. Gaga deve receber o dobro da quantidade que cada membro recebe.

**Problema:**

Crie um programa que determine a quantidade de fichas que Gaga recebeu, dado o número total de fichas no saco e o número de membros da equipe (excluindo a própria Gaga).

**Entrada:**

* A primeira linha contém um número inteiro `total_fichas`, que representa o número total de fichas.
* A segunda linha contém um número inteiro `numero_membros`, que representa a quantidade de membros da equipe (excluindo Gaga).

```python
total_fichas = int(input("Digite o total de fichas (jóias/drogas): "))
numero_membros = int(input("Digite o número de membros da equipe: "))

# A divisão considera 1 parte para cada membro e 2 partes para Gaga, totalizando (numero_membros + 2) partes.
fichas_membros = total_fichas // (numero_membros + 2)
fichas_gaga = 2 * fichas_membros

print(f"Gaga recebeu {fichas_gaga} fichas.")
```
